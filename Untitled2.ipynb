````python
import numpy as np
import matplotlib.pyplot as plt

# =============================================================================
# SECTION 1: SYSTEM CONSTANTS AND INITIAL CONDITIONS
# =============================================================================

# System Parameters
alpha = 1.0      # Cyclotron frequency alpha = qB/m (used in ESR evolution)
omega_B = alpha  # Convenience variable for magnetic field calculations
B_field = 1.0    # Magnetic Field magnitude (used in Region 2 velocity drift calc)
E_field = 1.5    # Magnitude of the Electric Field in Region II
alpha_angle = np.pi / 2 # E-field angle relative to x-axis
T_prime = -10.0  # Start of E-field pulse (Region I ends, B-field only)
T = 4.0          # End of E-field pulse (Region II ends, B-field only resumes)

# Initial conditions for the central trajectory at t=T' (The start of the calculation)
# We set the initial state (position and velocity) at the first boundary T_prime
x0_Tprime, y0_Tprime = 2.0, 2.0
vx0_Tprime, vy0_Tprime = 2.0, -2.0

Delta = 0.1  # Initial separation magnitude

# Initial separation vectors for the 4 corners
Dx0 = np.array([-Delta, Delta, Delta, -Delta])
Dy0 = np.array([-Delta, -Delta, Delta, Delta])


# =============================================================================
# SECTION 2: EQUATIONS OF MOTION FOR EACH REGION (USER PROVIDED)
# =============================================================================

def calculate_state_reg1(t, x0, y0, vx0, vy0):
    """
    Calculates state in Region I (B-field, t <= T_prime).
    Initial conditions are at t=T' (x0, y0, vx0, vy0).
    """
    angle = omega_B * (t - T_prime)
    sin_a = np.sin(angle)
    cos_a = np.cos(angle)

    vx = vx0 * cos_a + vy0 * sin_a
    vy = -vx0 * sin_a + vy0 * cos_a

    # Position equations exactly as provided in the prompt's Reg1 function
    x = x0 + (vx0 / omega_B) * sin_a - (vy0 / omega_B) * (1.0 + cos_a)
    y = y0 - (vx0 / omega_B) * (1.0 - cos_a) + (vy0 / omega_B) * sin_a

    return x, y, vx, vy

def calculate_state_reg2(t, x0_II, y0_II, ux0_II, uy0_II):
    """
    Calculates state in Region II (E & B fields, T_prime < t < T).
    Initial conditions are at t=0 (x0_II, y0_II, ux0_II, uy0_II).
    """
    # Note: These equations use t relative to t=0, so the initial condition (x0_II...) must be the state at t=0.

    v_drift_x = E_field / B_field * np.sin(alpha_angle)
    v_drift_y = -E_field / B_field * np.cos(alpha_angle)

    common_factor_x = (ux0_II / omega_B - v_drift_x / omega_B)
    common_factor_y = (uy0_II / omega_B - v_drift_y / omega_B)

    # Position equations exactly as provided in the prompt's Reg2 function
    gyration_x = common_factor_x * np.sin(omega_B * t) - common_factor_y * (np.cos(omega_B * t) - 1)
    gyration_y = common_factor_x * (np.cos(omega_B * t) - 1) + common_factor_y * np.sin(omega_B * t)
    x = x0_II + v_drift_x * t + gyration_x
    y = y0_II + v_drift_y * t + gyration_y

    # Velocity equations exactly as provided in the prompt's Reg2 function
    ux = v_drift_x + (ux0_II - v_drift_x) * np.cos(omega_B * t) + (uy0_II - v_drift_y) * np.sin(omega_B * t)
    uy = v_drift_y - (ux0_II - v_drift_x) * np.sin(omega_B * t) + (uy0_II - v_drift_y) * np.cos(omega_B * t)

    return x, y, ux, uy

def calculate_state_reg3(t, x0_III, y0_III, vx0_III, vy0_III):
    """
    Calculates state in Region III (B-field, t >= T).
    Initial conditions are at t=T (x0_III, y0_III, vx0_III, vy0_III).
    """
    t_rel = t - T
    angle = omega_B * t_rel
    sin_a = np.sin(angle)
    cos_a = np.cos(angle)

    vx = vx0_III * cos_a + vy0_III * sin_a
    vy = -vx0_III * sin_a + vy0_III * cos_a

    # Position equations exactly as provided in the prompt's Reg3 function
    x = x0_III + (vy0_III / omega_B) + (vx0_III / omega_B) * sin_a - (vy0_III / omega_B) * cos_a
    y = y0_III - (vx0_III / omega_B) + (vx0_III / omega_B) * cos_a + (vy0_III / omega_B) * sin_a

    return x, y, vx, vy

# =============================================================================
# SECTION 3: ESR EVOLUTION (REGION-BY-REGION WITH CONTINUITY)
# =============================================================================

def esr_region1(t_array, theta0, sigma_plus0, sigma_x0, omega0):
    """
    Calculates ESR evolution in Region I (B-field only, t <= 0).
    Returns full evolution arrays AND final values at t=0 (boundary with Region II).

    # MODIFIED: Added return of final ESR values for Region II initialization
    """
    theta_array = np.zeros_like(t_array, dtype=float)
    sigma_plus_array = np.zeros_like(t_array, dtype=float)
    sigma_x_array = np.zeros_like(t_array, dtype=float)
    omega_array = np.zeros_like(t_array, dtype=float)

    C_raw = (1.0 / alpha) * np.arctan(theta0 / alpha)
    C = C_raw + np.pi / alpha

    den_c = 1.0 + (theta0 / alpha)**2
    E = sigma_plus0 / den_c
    F = sigma_x0 / den_c
    G = (omega0 - alpha / 2) / den_c

    for idx, t in enumerate(t_array):
        arg = alpha * (C - t / 2)
        theta_array[idx] = alpha * np.tan(arg)
        S = 1.0 / (np.cos(arg))**2
        sigma_plus_array[idx] = S * (E * np.sin(alpha * t) + F * np.cos(alpha * t))
        sigma_x_array[idx] = S * (E * np.cos(alpha * t) - F * np.sin(alpha * t))
        omega_array[idx] = alpha / 2 + G * S

    # MODIFIED: Extract and return final values at t=0 boundary
    arg_boundary = alpha * (C - 0.0 / 2)
    S_boundary = 1.0 / (np.cos(arg_boundary))**2
    theta_boundary = alpha * np.tan(arg_boundary)
    sigma_plus_boundary = S_boundary * (E * np.sin(0.0) + F * np.cos(0.0))
    sigma_x_boundary = S_boundary * (E * np.cos(0.0) - F * np.sin(0.0))
    omega_boundary = alpha / 2 + G * S_boundary

    return t_array, theta_array, sigma_plus_array, sigma_x_array, omega_array, \
           (theta_boundary, sigma_plus_boundary, sigma_x_boundary, omega_boundary)


def esr_region2(t_array, theta0_r2, sigma_plus0_r2, sigma_x0_r2, omega0_r2):
    """
    Calculates ESR evolution in Region II (E & B fields, 0 < t <= T).
    Takes initial ESR values from Region I at t=0.
    Returns full evolution arrays AND final values at t=T (boundary with Region III).

    # MODIFIED: Accepts initial conditions from Region I, returns boundary values for Region III
    """
    theta_array = np.zeros_like(t_array, dtype=float)
    sigma_plus_array = np.zeros_like(t_array, dtype=float)
    sigma_x_array = np.zeros_like(t_array, dtype=float)
    omega_array = np.zeros_like(t_array, dtype=float)

    C_raw = (1.0 / alpha) * np.arctan(theta0_r2 / alpha)
    C = C_raw + np.pi / alpha

    den_c = 1.0 + (theta0_r2 / alpha)**2
    E = sigma_plus0_r2 / den_c
    F = sigma_x0_r2 / den_c
    G = (omega0_r2 - alpha / 2) / den_c

    for idx, t in enumerate(t_array):
        arg = alpha * (C - t / 2)
        theta_array[idx] = alpha * np.tan(arg)
        S = 1.0 / (np.cos(arg))**2
        sigma_plus_array[idx] = S * (E * np.sin(alpha * t) + F * np.cos(alpha * t))
        sigma_x_array[idx] = S * (E * np.cos(alpha * t) - F * np.sin(alpha * t))
        omega_array[idx] = alpha / 2 + G * S

    # MODIFIED: Extract and return final values at t=T boundary
    arg_boundary = alpha * (C - T / 2)
    S_boundary = 1.0 / (np.cos(arg_boundary))**2
    theta_boundary = alpha * np.tan(arg_boundary)
    sigma_plus_boundary = S_boundary * (E * np.sin(alpha * T) + F * np.cos(alpha * T))
    sigma_x_boundary = S_boundary * (E * np.cos(alpha * T) - F * np.sin(alpha * T))
    omega_boundary = alpha / 2 + G * S_boundary

    return t_array, theta_array, sigma_plus_array, sigma_x_array, omega_array, \
           (theta_boundary, sigma_plus_boundary, sigma_x_boundary, omega_boundary)


def esr_region3(t_array, theta0_r3, sigma_plus0_r3, sigma_x0_r3, omega0_r3):
    """
    Calculates ESR evolution in Region III (B-field only, t > T).
    Takes initial ESR values from Region II at t=T.

    # MODIFIED: Accepts initial conditions from Region II, completes evolution chain
    """
    theta_array = np.zeros_like(t_array, dtype=float)
    sigma_plus_array = np.zeros_like(t_array, dtype=float)
    sigma_x_array = np.zeros_like(t_array, dtype=float)
    omega_array = np.zeros_like(t_array, dtype=float)

    C_raw = (1.0 / alpha) * np.arctan(theta0_r3 / alpha)
    C = C_raw + np.pi / alpha

    den_c = 1.0 + (theta0_r3 / alpha)**2
    E = sigma_plus0_r3 / den_c
    F = sigma_x0_r3 / den_c
    G = (omega0_r3 - alpha / 2) / den_c

    for idx, t in enumerate(t_array):
        t_rel = t - T  # Time relative to Region III start
        arg = alpha * (C - t_rel / 2)
        theta_array[idx] = alpha * np.tan(arg)
        S = 1.0 / (np.cos(arg))**2
        sigma_plus_array[idx] = S * (E * np.sin(alpha * t_rel) + F * np.cos(alpha * t_rel))
        sigma_x_array[idx] = S * (E * np.cos(alpha * t_rel) - F * np.sin(alpha * t_rel))
        omega_array[idx] = alpha / 2 + G * S

    return t_array, theta_array, sigma_plus_array, sigma_x_array, omega_array

# =============================================================================
# SECTION 4: COMPOSITE TRAJECTORY (HANDLING BOUNDARIES)
# =============================================================================

def calculate_trajectory_composite(t, x0_Tprime, y0_Tprime, vx0_Tprime, vy0_Tprime):
    """
    Calculates the full trajectory across the three time regions.
    The start time for the input state is T_prime = -10.0.
    """

    x = np.zeros_like(t, dtype=float)
    y = np.zeros_like(t, dtype=float)
    vx = np.zeros_like(t, dtype=float)
    vy = np.zeros_like(t, dtype=float)

    # 1. Calculate State at Boundary 2 (t=0)
    # Region 1 runs from T_prime to 0. The inputs are the state at T_prime.
    x0_II, y0_II, vx0_II, vy0_II = calculate_state_reg1(0.0, x0_Tprime, y0_Tprime, vx0_Tprime, vy0_Tprime)

    # 2. Calculate State at Boundary 3 (t=T)
    # Region 2 runs from 0 to T. The inputs are the state at t=0.
    x0_III, y0_III, vx0_III, vy0_III = calculate_state_reg2(T, x0_II, y0_II, vx0_II, vy0_II)

    for idx, time in enumerate(t):
        if time <= T_prime:
            # Region I (t < T_prime): B-field only, using state at T' as reference
            x[idx], y[idx], vx[idx], vy[idx] = calculate_state_reg1(time, x0_Tprime, y0_Tprime, vx0_Tprime, vy0_Tprime)
        elif time <= 0.0:
            # Region I (T' < t <= 0): B-field only, using state at T' as reference
            x[idx], y[idx], vx[idx], vy[idx] = calculate_state_reg1(time, x0_Tprime, y0_Tprime, vx0_Tprime, vy0_Tprime)
        elif time <= T:
            # Region II (0 < t <= T): E & B fields, using state at t=0 as reference
            x[idx], y[idx], vx[idx], vy[idx] = calculate_state_reg2(time, x0_II, y0_II, vx0_II, vy0_II)
        else:
            # Region III (t > T): B-field only, using state at t=T as reference
            x[idx], y[idx], vx[idx], vy[idx] = calculate_state_reg3(time, x0_III, y0_III, vx0_III, vy0_III)

    return x, y, vx, vy

# =============================================================================
# SECTION 5: INITIAL VELOCITY SETUP & PLOTTING DEMONSTRATION
# =============================================================================

def get_initial_velocity_family(theta0, sigma_plus0, sigma_x0, omega0):
    """
    Calculates initial velocities for the 4 corner trajectories at t=T' (Eq. 3.11).
    """
    A11 = theta0 / 2 + sigma_plus0
    A12 = sigma_x0 + omega0
    A21 = sigma_x0 - omega0
    A22 = theta0 / 2 - sigma_plus0

    Dux0 = A11 * Dx0 + A12 * Dy0
    Duy0 = A21 * Dx0 + A22 * Dy0

    # u_i(T') = u_center(T') + Du_0
    ux0_i = vx0_Tprime + Dux0
    uy0_i = vy0_Tprime + Duy0

    return ux0_i, uy0_i

# Example ESR Parameters (I=0, focusing)
theta0_ex = 2.0
sigma_plus0_ex = 0.3
sigma_x0_ex = 0.4
omega0_ex = 0.0

# Calculate plot range
t_f_ref = 2 * ((1.0 / alpha) * np.arctan(theta0_ex / alpha) + np.pi / alpha)
t_end_plot = 15.0
t_start_plot = T_prime

# Create time arrays for each region
t_region1 = np.linspace(t_start_plot, 0.0, 100)
t_region2 = np.linspace(0.0, T, 100)
t_region3 = np.linspace(T, t_end_plot, 100)

# MODIFIED: Run ESR through Region I and extract boundary values
_, theta_r1, sigma_plus_r1, sigma_x_r1, omega_r1, boundary_r1 = \
    esr_region1(t_region1, theta0_ex, sigma_plus0_ex, sigma_x0_ex, omega0_ex)

# MODIFIED: Pass Region I boundary values as initial conditions to Region II
theta_r1_boundary, sigma_plus_r1_boundary, sigma_x_r1_boundary, omega_r1_boundary = boundary_r1
_, theta_r2, sigma_plus_r2, sigma_x_r2, omega_r2, boundary_r2 = \
    esr_region2(t_region2, theta_r1_boundary, sigma_plus_r1_boundary, sigma_x_r1_boundary, omega_r1_boundary)

# MODIFIED: Pass Region II boundary values as initial conditions to Region III
theta_r2_boundary, sigma_plus_r2_boundary, sigma_x_r2_boundary, omega_r2_boundary = boundary_r2
_, theta_r3, sigma_plus_r3, sigma_x_r3, omega_r3 = \
    esr_region3(t_region3, theta_r2_boundary, sigma_plus_r2_boundary, sigma_x_r2_boundary, omega_r2_boundary)

# Concatenate all regions (avoiding duplicates at boundaries)
t_esr = np.concatenate([t_region1, t_region2[1:], t_region3[1:]])
theta_esr = np.concatenate([theta_r1, theta_r2[1:], theta_r3[1:]])
sigma_plus_esr = np.concatenate([sigma_plus_r1, sigma_plus_r2[1:], sigma_plus_r3[1:]])
sigma_x_esr = np.concatenate([sigma_x_r1, sigma_x_r2[1:], sigma_x_r3[1:]])
omega_esr = np.concatenate([omega_r1, omega_r2[1:], omega_r3[1:]])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# --- Left Plot: Trajectories ---
ax1.set_title(r'Pulsed Cyclotron Trajectories ($\mathbf{E}$ pulse $t \in [0, %s]$)' % T)
ax1.set_xlabel(r'x-position')
ax1.set_ylabel(r'y-position')

ax1.axvline(x=T_prime, color='orange', linestyle='--', label=r'Start of Simulation ($t=%s$)' % T_prime)
ax1.axvline(x=0.0, color='blue', linestyle='--', label=r'E-field ON ($t=0$)')
ax1.axvline(x=T, color='red', linestyle='--', label=r'E-field OFF ($t=%s$)' % T)

# Central Trajectory
t_plot_traj = np.linspace(t_start_plot, t_end_plot, 500)
x_c, y_c, _, _ = calculate_trajectory_composite(t_plot_traj, x0_Tprime, y0_Tprime, vx0_Tprime, vy0_Tprime)
ax1.plot(x_c, y_c, color='black', label='Center Trajectory', linewidth=1.5)

# Family Trajectories
ux0_i, uy0_i = get_initial_velocity_family(theta0_ex, sigma_plus0_ex, sigma_x0_ex, omega0_ex)
for i in range(4):
    x_i, y_i, _, _ = calculate_trajectory_composite(t_plot_traj, x0_Tprime + Dx0[i], y0_Tprime + Dy0[i], ux0_i[i], uy0_i[i])
    ax1.plot(x_i, y_i, linewidth=0.8, linestyle='--')

ax1.legend()
ax1.set_aspect('equal')

# --- Right Plot: ESR Evolution Through Regions ---
ax2.set_title(r'Continuous ESR Evolution Through Regions ($\theta_0=%s$ at $t=%s$)' % (theta0_ex, T_prime))
ax2.axvline(T_prime, color='orange', linestyle='--', label=r'Start ($t=%s$)' % T_prime)
ax2.axvline(0.0, color='blue', linestyle='--', label=r'E-field ON ($t=0$)')
ax2.axvline(T, color='red', linestyle='--', label=r'E-field OFF ($t=%s$)' % T)

ax2.plot(t_esr, theta_esr, label=r'$\theta$ (Expansion)', color='tab:blue')
ax2.plot(t_esr, sigma_plus_esr, label=r'$\sigma_+$ (Shear)', color='tab:red')
ax2.plot(t_esr, sigma_x_esr, label=r'$\sigma_\times$ (Shear)', color='tab:green')
ax2.plot(t_esr, omega_esr, label=r'$\omega$ (Rotation)', color='tab:purple')

ax2.set_xlabel(r'Time $t$')
ax2.set_ylabel(r'Kinematic Variables ($\theta, \sigma, \omega$)')
ax2.axhline(0, color='k', linestyle=':', linewidth=0.5)
ax2.set_ylim(-10, 5)
ax2.legend(loc='lower left')

plt.tight_layout()
plt.savefig('pulsed_cyclotron_kinematics_t_prime_neg10.png')
print("pulsed_cyclotron_kinematics_t_prime_neg10.png")
````

````python
print("\n--- Calculating Critical Times for Theta to Tend to -Infinity ---\n")

def calculate_esr_C(theta_initial, alpha_val):
    """Calculates the C constant for a given initial theta and alpha."""
    return (1.0 / alpha_val) * np.arctan(theta_initial / alpha_val) + np.pi / alpha_val

def calculate_t_relative_star(C_val, alpha_val, n_val):
    """Calculates the relative time t_star where theta tends to -infinity."""
    return 2 * C_val - (2 * n_val + 1) * np.pi / alpha_val


critical_times_region1 = []
critical_times_region2 = []
critical_times_region3 = []

# --- Region I Calculations ---
print(f"Region I (Time range: {T_prime:.1f} to 0.0)")
theta0_r1_calc = theta0_ex
C_r1_calc = calculate_esr_C(theta0_r1_calc, alpha)

# Calculate possible n values. The argument starts at alpha*C and decreases.
# We want alpha * (C - t_rel/2) = (2n+1)*pi/2
# For theta to go to -infinity, arg must approach (2n+1)pi/2 from positive side
# This means arg_initial > (2n+1)pi/2
max_n_r1 = int(np.floor(alpha * C_r1_calc / np.pi - 0.5)) # Max n for which (2n+1)pi/2 < alpha*C
min_n_r1 = int(np.ceil(alpha * C_r1_calc / np.pi - (T_prime / 2 + 0.5))) # Min n for t_relative_star > T_prime. This is a bit tricky, just use a range for n.

# Iterate through a reasonable range of n values (e.g., -5 to 5 or based on max_n_r1)
for n in range(max_n_r1, -5, -1): # Iterate downwards to find relevant singularities first
    t_relative_star = calculate_t_relative_star(C_r1_calc, alpha, n)
    # Check if t_relative_star is within the region's relative time range [0, -T_prime]
    if 0 <= t_relative_star <= (0.0 - T_prime):
        t_actual_star = T_prime + t_relative_star
        critical_times_region1.append(t_actual_star)
        print(f"  - Critical time at t = {t_actual_star:.4f} (relative t = {t_relative_star:.4f}) for n = {n}")

# --- Region II Calculations ---
print(f"\nRegion II (Time range: 0.0 to {T:.1f})")
theta0_r2_calc = theta_r1_boundary
C_r2_calc = calculate_esr_C(theta0_r2_calc, alpha)

max_n_r2 = int(np.floor(alpha * C_r2_calc / np.pi - 0.5))
for n in range(max_n_r2, -5, -1):
    t_relative_star = calculate_t_relative_star(C_r2_calc, alpha, n)
    # Check if t_relative_star is within the region's relative time range [0, T]
    if 0 <= t_relative_star <= T:
        t_actual_star = 0.0 + t_relative_star
        critical_times_region2.append(t_actual_star)
        print(f"  - Critical time at t = {t_actual_star:.4f} (relative t = {t_relative_star:.4f}) for n = {n}")

# --- Region III Calculations ---
print(f"\nRegion III (Time range: {T:.1f} to {t_end_plot:.1f})")
theta0_r3_calc = theta_r2_boundary
C_r3_calc = calculate_esr_C(theta0_r3_calc, alpha)

max_n_r3 = int(np.floor(alpha * C_r3_calc / np.pi - 0.5))
for n in range(max_n_r3, -5, -1):
    t_relative_star = calculate_t_relative_star(C_r3_calc, alpha, n)
    # Check if t_relative_star is within the region's relative time range [0, t_end_plot - T]
    if 0 <= t_relative_star <= (t_end_plot - T):
        t_actual_star = T + t_relative_star
        critical_times_region3.append(t_actual_star)
        print(f"  - Critical time at t = {t_actual_star:.4f} (relative t = {t_relative_star:.4f}) for n = {n}")
````

## Condition for $\theta$ to tend to negative infinity

The ESR variable $\theta(t)$ tends to negative infinity when the argument of the tangent function approaches an odd multiple of $\frac{\pi}{2}$ from the positive side. Since $\alpha > 0$ and $t_{relative}$ (time relative to the start of each region) increases, the argument $\alpha \left(C - \frac{t_{relative}}{2}\right)$ is a decreasing function.

Therefore, $\theta(t)$ will tend to negative infinity as $t_{relative}$ approaches a critical time $t^*$ (from values less than $t^*$) such that:

$$\alpha \left(C - \frac{t^*}{2}\right) = \frac{(2n+1)\pi}{2} \quad \text{for some integer } n$$

Substituting the expression for $C$, this critical time $t^*$ is given by:

$$t^* = \frac{2}{\alpha} \arctan\left(\frac{\theta_0}{\alpha}\right) + \frac{\pi}{\alpha}(1 - 2n)$$

where $\theta_0$ is the initial $\theta$ value for the specific region being considered. This condition indicates a singularity or resonance in the system.

The ESR variable $\theta$ is calculated using the following equation in all three regions, with `t` being the time relative to the start of each region, and `C` determined by the initial conditions of that region:

$$\theta(t) = \alpha \tan\left(\alpha \left(C - \frac{t}{2}\right)\right)$$

where:

*   $\alpha$ is the cyclotron frequency.
*   $C = \frac{1}{\alpha} \arctan\left(\frac{\theta_0}{\alpha}\right) + \frac{\pi}{\alpha}$, where $\theta_0$ is the initial value of $\theta$ for that region.
*   For Region I, $t$ is relative to $T_{prime}$. The initial $\theta_0$ is `theta0` (e.g., `theta0_ex`).
*   For Region II, $t$ is relative to $0$. The initial $\theta_0$ is the value of $\theta$ at $t=0$ from Region I.
*   For Region III, $t$ is relative to $T$. The initial $\theta_0$ is the value of $\theta$ at $t=T$ from Region II.

def calculate_state_reg1(t, x0, y0, vx0, vy0):
    """
    Calculates state in Region I (B-field, t <= T_prime).
    Initial conditions are at t=T' (x0, y0, vx0, vy0).
    """
    angle = omega_B * (t - T_prime)
    sin_a = np.sin(angle)
    cos_a = np.cos(angle)

    vx = vx0 * cos_a + vy0 * sin_a
    vy = -vx0 * sin_a + vy0 * cos_a

    # Position equations exactly as provided in the prompt's Reg1 function
    x = x0 + (vx0 / omega_B) * sin_a - (vy0 / omega_B) * (1.0 + cos_a)
    y = y0 - (vx0 / omega_B) * (1.0 - cos_a) + (vy0 / omega_B) * sin_a

    return x, y, vx, vy

def calculate_state_reg2(t, x0_II, y0_II, ux0_II, uy0_II):
    """
    Calculates state in Region II (E & B fields, T_prime < t < T).
    Initial conditions are at t=0 (x0_II, y0_II, ux0_II, uy0_II).
    """
    # Note: These equations use t relative to t=0, so the initial condition (x0_II...) must be the state at t=0.

    v_drift_x = E_field / B_field * np.sin(alpha_angle)
    v_drift_y = -E_field / B_field * np.cos(alpha_angle)

    common_factor_x = (ux0_II / omega_B - v_drift_x / omega_B)
    common_factor_y = (uy0_II / omega_B - v_drift_y / omega_B)

    # Position equations exactly as provided in the prompt's Reg2 function
    gyration_x = common_factor_x * np.sin(omega_B * t) - common_factor_y * (np.cos(omega_B * t) - 1)
    gyration_y = common_factor_x * (np.cos(omega_B * t) - 1) + common_factor_y * np.sin(omega_B * t)
    x = x0_II + v_drift_x * t + gyration_x
    y = y0_II + v_drift_y * t + gyration_y

    # Velocity equations exactly as provided in the prompt's Reg2 function
    ux = v_drift_x + (ux0_II - v_drift_x) * np.cos(omega_B * t) + (uy0_II - v_drift_y) * np.sin(omega_B * t)
    uy = v_drift_y - (ux0_II - v_drift_x) * np.sin(omega_B * t) + (uy0_II - v_drift_y) * np.cos(omega_B * t)

    return x, y, ux, uy

def calculate_state_reg3(t, x0_III, y0_III, vx0_III, vy0_III):
    """
    Calculates state in Region III (B-field, t >= T).
    Initial conditions are at t=T (x0_III, y0_III, vx0_III, vy0_III).

    vx = vx0_III * cos_a + vy0_III * sin_a

    x = x0_III + (vy0_III / omega_B) + (vx0_III / omega_B) * sin_a - (vy0_III / omega_B) * cos_a
    y = y0_III - (vx0_III / omega_B) + (vx0_III / omega_B) * cos_a + (vy0_III / omega_B) * sin_a

    return x, y, vx, vy

def esr_region1(t_array, theta0, sigma_plus0, sigma_x0, omega0):
    """
    Calculates ESR evolution in Region I (B-field only, t <= 0).
    Returns full evolution arrays AND final values at t=0 (boundary with Region II).
    """
    theta_array = np.zeros_like(t_array, dtype=float)
    sigma_plus_array = np.zeros_like(t_array, dtype=float)
    sigma_x_array = np.zeros_like(t_array, dtype=float)
    omega_array = np.zeros_like(t_array, dtype=float)

    C_raw = (1.0 / alpha) * np.arctan(theta0 / alpha)
    C = C_raw + np.pi / alpha

    den_c = 1.0 + (theta0 / alpha)**2
    E = sigma_plus0 / den_c
    F = sigma_x0 / den_c
    G = (omega0 - alpha / 2) / den_c

    for idx, t in enumerate(t_array):
        arg = alpha * (C - t / 2)
        theta_array[idx] = alpha * np.tan(arg)
        S = 1.0 / (np.cos(arg))**2
        sigma_plus_array[idx] = S * (E * np.sin(alpha * t) + F * np.cos(alpha * t))
        sigma_x_array[idx] = S * (E * np.cos(alpha * t) - F * np.sin(alpha * t))
        omega_array[idx] = alpha / 2 + G * S

    # Extract and return final values at t=0 boundary
    arg_boundary = alpha * (C - 0.0 / 2)
    S_boundary = 1.0 / (np.cos(arg_boundary))**2
    theta_boundary = alpha * np.tan(arg_boundary)
    sigma_plus_boundary = S_boundary * (E * np.sin(0.0) + F * np.cos(0.0))
    sigma_x_boundary = S_boundary * (E * np.cos(0.0) - F * np.sin(0.0))
    omega_boundary = alpha / 2 + G * S_boundary

    return t_array, theta_array, sigma_plus_array, sigma_x_array, omega_array, \
           (theta_boundary, sigma_plus_boundary, sigma_x_boundary, omega_boundary)

def calculate_trajectory_composite(t, x0_Tprime, y0_Tprime, vx0_Tprime, vy0_Tprime):

    x0_II, y0_II, vx0_II, vy0_II = calculate_state_reg1(0.0, x0_Tprime, y0_Tprime, vx0_Tprime, vy0_Tprime)

    # 2. Calculate State at Boundary 3 (t=T)

        elif time <= 0.0:
